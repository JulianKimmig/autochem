```python
import pandas as pd
import numpy as np
import os
```

```python
MEASURMENT_FILE="measurements.xlsx"
RRF_FILE="RRF_values.xlsx"
```

```python
def do_eval_time(df,rrfs):
    assert len(df["time"].unique())==1
    # bring all empty measuremtns to zero
    for r,d in df.iterrows():
        rrf=rrfs[d["RRF"]]
        for k in rrf.keys():
            if k in d:
                df.loc[r,k]=np.nan_to_num(df.loc[r,k])
    display(df)
    
def do_auto_eval_all(df,rrfs):
    dfs = [x for _, x in df.groupby('time')]
    for df in dfs:
        do_eval_time(df,rrfs)


rrfs=pd.read_excel(RRF_FILE,index_col=0).to_dict()
        
with pd.ExcelFile(MEASURMENT_FILE) as excel_reader:
    for s in excel_reader.sheet_names:
        do_auto_eval_all(pd.read_excel(MEASURMENT_FILE,sheet_name=s,header=None,index_col=0).transpose(),rrfs)
```
